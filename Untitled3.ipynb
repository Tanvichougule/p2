```python
#intersection and difference
import numpy as np
n1=np.array([10,20,30,40,50,60])
n2=np.array([70,80,90,])
np.intersect1d(n1,n2)
```

```python
import numpy as np
n1=np.array([10,20,30,40,50,60])
n2=np.array([70,80,90,])
np.1d(n1,n2)
```

```python
import numpy as np
n1=np.array([10,20,30,40,50,60])
n2=np.array([70,80,90,])
np.setdiff1d(n2,n1)
```
